```python
df_samsung = df[['ds','Open']].rename(axis=1,mapper={'Open':'y','Date':'ds'})
df_samsung['ds'] = pd.to_datetime(df_samsung['ds'])
```

```python
tr1 = go.Scatter(x=df_samsung.ds,y = df_samsung.y)
iplot([tr1])
```

```python
df_test  = df_samsung.iloc[:-15]
m = Prophet(yearly_seasonality=False,
            weekly_seasonality=False)


m.fit(df_test)

future = m.make_future_dataframe(periods=15,freq='D',)
forecast = m.predict(future)
df_results = pd.merge(how='left',left=df_samsung,right=forecast,on='ds')
```
